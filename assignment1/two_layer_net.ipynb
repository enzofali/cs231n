## Inline Question 1:

We've only asked you to implement ReLU, but there are a number of different activation functions that one could use in neural networks, each with its pros and cons. In particular, an issue commonly seen with activation functions is getting zero (or close to zero) gradient flow during backpropagation. Which of the following activation functions have this problem? If you consider these functions in the one dimensional case, what types of input would lead to this behaviour?
1. Sigmoid
2. ReLU
3. Leaky ReLU

## Answer:

El problema del "desvanecimiento de gradientes" se presenta cuando los gradientes de las funciones de activación son muy pequeños, lo que dificulta el entrenamiento de las redes al actualizar el gradiente con cambios insignificantes.

1. Sigmoid

La derivada de la función de activación sigmoide tiende a aproximarse a cero cuando los valores de entrada son muy grandes, ya sean positivos o negativos. Por lo tanto, los valores de entrada no normalizados pueden resultar en gradientes pequeños, induciendo el problema mencionado.

![image.png](https://miro.medium.com/v2/resize:fit:330/format:webp/1*0yhJ7DbhOX-tRUseljjYoA.png)

2. ReLU (Rectified Linear Unit)

La derivada de la función de activación ReLU es `0` cuando `x < 0` y `1` cuando `x > 0`. Debido a que mantiene un gradiente constante, independiente de los valores de entrada, ReLU no sufre del problema de desvanecimiento de gradientes. Sin embargo, para entradas negativas, el gradiente es igual a cero, lo que puede generar un problema conocido como "dying ReLU", donde las neuronas se vuelven inactivas y dejan de aprender, ya que el gradiente es cero en ese rango.

3. Leaky ReLU

Leaky ReLU es una variante de la ReLU que permite un pequeño gradiente no nulo para las entradas negativas, abordando en el problema de desvanecimiento de gradientes. Por lo tanto, esta variante no sufre del problema de desvanecimiento de gradientes.

$\color{blue}{\textit Your Explanation:}$

1. Entrenar con un conjunto de datos más grande

Aumentar el tamaño del conjunto de entrenamiento puede mejorar la capacidad de generalización del modelo, disminuyendo así el sobreajuste, el cual genera la diferencia entre la precisión en el conjunto de entrenamiento y en el conjunto de prueba.

2. Añadir más unidades ocultas.

Como se planteó en la sección anterior, al añadir más unidades ocultas (aumentar el tamaño del modelo), esperaríamos ver un mayor sobreajuste de nuestros datos, lo cual se manifiesta como una mayor diferencia entre la precisión en el entrenamiento y la precisión en la validación.

3. Aumentar la fuerza de regularización.

El propósito de las técnicas de regularización es prevenir el sobreajuste mediante la inclusión de un término de penalización en la función de pérdida, lo que reduce la diferencia entre las precisiones en los conjuntos.

Por ejemplo, L1 desplaza el valor de muchos pesos hacia 0, fomenta la selección de características y reduce la complejidad del modelo (mayor interpretabilidad). L2 distribuye la penalización entre todos los pesos, evitando la dependencia excesiva del modelo a una característica particular y previniendo el sobreajuste cuando se tiene un gran número de características que son potencialmente relevantes.
